# Создаю таблицы

```python
test_1 = pd.read_excel('Тестовое задание Excel КР.xlsx', sheet_name='Исх к заданиям', skiprows=1, index_col ='Табельный номер') # Считываю 1-й лист
```

# Заполнить таблицу 4 данными из таблицы 1.

```python
table_4 = pd.read_excel('Тестовое задание Excel КР.xlsx', sheet_name='Условия к заданию 1', skiprows=1) # считываю 2-й лист
```

```python
table_4 = table_4[:94].set_index('Табельный номер') # Привожу табельный номер к индексу
```

```python
table_4.index = table_4.index.astype('int64') # Привожу индекс в порядок(записался типа float)
```

```python
pivot_table_1 = pd.pivot_table(table_1, index = table_1['ФИО'], values='Сумма', aggfunc='sum') # сводная таблица для добавления данных
```

```python
tab_num = table_1.reset_index().set_index('ФИО')['Табельный номер'] # достаю данные из 1 -й таблицы для присваивания табельного номера
```

```python
pivot_table_1 = pivot_table_1.merge(tab_num, left_on=pivot_table_1.index, right_on=tab_num.index, how='left') # присваиваю по ФИО табельный номер
```

```python
pivot_table_1 = pivot_table_1.drop_duplicates().set_index('Табельный номер').rename(columns= {'key_0' : 'ФИО'}) # убираю дубликаты и переминовываю колонку
```

```python
data_table_FIO = dict(zip(pivot_table_1.index, pivot_table_1['ФИО'])) # создаю словарь колнки "ФИО"
```

```python
data_table_SUM = dict(zip(pivot_table_1.index, pivot_table_1['Сумма'])) # создаю словарь колнки "Сумма"
```

```python
def add_in_test_2(table_fill_in, data, column_name): 
    for i in table_fill_in.index:   # прохожусь по индексам, сравниваю значения и добавляю значения из нужных колонок
        for key, value in data.items():
            if i == key:
                table_fill_in.loc[i, column_name] = value
```

```python
add_in_test_2(table_4, data_table_FIO, 'ФИО')
add_in_test_2(table_4, data_table_SUM, 'Сумма')
```

```python
table_4 # Результат
```

# В случае, если по табельному номеру нет сотрудника, укажите в формуле условие, чтоб в столбце ФИО формула возвращала "Нет табельного", а в столбце сумма "0".

```python
table_4['ФИО'] = table_4['ФИО'].fillna('Нет табельного') # Заполняю пустые ячейки
```

```python
table_4['Сумма'] = table_4['Сумма'].fillna(0).astype('int64') # Заполняю пустые ячейки, меняю тип 
```

```python
table_4 # результат
```

# Проставьте в таблице 4 статусы, возьмите их в таблице 2.

```python
table_4['Статус'] = table_4['Статус'].astype('object').fillna('-') # Меняю тип и заполняю пустые ячейки 
```

```python
status_A = table_4[table_4['Сумма'] > 30000] # Достаю значения больше 30000
```

```python
status_B = table_4[                                       # Достаю значения в диапазоне от 25000 до 30000
                          (table_4['Сумма'] < 30000) &
                          (table_4['Сумма'] > 25000)
    ]
```

```python
status_C = table_4[                                       # До 25000 
                          (table_4['Сумма'] < 25000) &
                          (table_4['Сумма'] > 0)
    ]
```

```python
def add_status(check_table, column_name, status):      # создаю функцю для добавления нужного статуса в нужную ячейку
    for id_1 in check_table.index:
        for id_2 in column_name.index:
            if id_1 == id_2:
                check_table.loc[id_1, 'Статус'] = status
```

```python
add_status(table_4, status_A, table_2['Статус'][1])  # добавляю нужные значения 
add_status(table_4, status_B, table_2['Статус'][2])
add_status(table_4, status_C, table_2['Статус'][3])
```

```python
table_4 # вывод результата
```
